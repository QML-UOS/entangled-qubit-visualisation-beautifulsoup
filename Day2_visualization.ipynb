```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

# Quantum speedup visualized

## Factoring

```python
x_values = np.array([0, 2, 2**2, 2**3, 2**4, 2**5, 2**6, 2**7, 2**8, 2**9, 2**10, 2**11])
shor_values = np.log(x_values ** 3)
classical_values = np.log(2 ** (x_values/4))

fig, ax = plt.subplots()
ax.plot(x_values, shor_values)
ax.plot(x_values, classical_values)
#plt.xticks(x_values)

ax.set(xlabel='Number of inputs N', ylabel='log-o-notation',
       title='Quantum speedup factoring')
ax.grid()

plt.show()
```
